```python
# ðŸ”¹ Carregar arquivo NetCDF (substitua pelo seu caminho no Colab ou Google Drive)
netcdf_path = '/content/drive/MyDrive/Movasc2_scripts/pr_20010101_20240320_BR-DWGD_UFES_UTEXAS_v_3.2.3.nc'
ds = xr.open_dataset(netcdf_path)

# ðŸ”¸ Supondo que seu NetCDF tenha variÃ¡veis de coordenadas chamadas 'lat' e 'lon'
netcdf_lats = ds['latitude'].values
netcdf_lons = ds['longitude'].values

# Criar uma matriz com todas as combinaÃ§Ãµes de lat/lon do NetCDF
netcdf_coords = np.array(np.meshgrid(netcdf_lats, netcdf_lons)).T.reshape(-1, 2)
netcdf_coords = np.array([[lat, lon] for lat, lon in netcdf_coords])
```

```python
# Para cada ponto de entrada, encontrar a coordenada mais prÃ³xima no NetCDF
closest_coords = []

for index, row in df_input.iterrows():
    point = np.array([row['lat'], row['lon']])

    # Calcula a distÃ¢ncia euclidiana entre o ponto e todas as coordenadas do NetCDF
    dists = distance.cdist([point], netcdf_coords, 'euclidean')

    # Ãndice da menor distÃ¢ncia
    min_idx = np.argmin(dists)

    # Coordenada mais prÃ³xima
    closest_point = netcdf_coords[min_idx]
    closest_coords.append(closest_point)

# Adicionando resultado ao GeoDataFrame
df_closest = pd.DataFrame(closest_coords, columns=['closest_lat', 'closest_lon'])

# Juntando dados
gdf_result = pd.concat([df_input, df_closest], axis=1)

# ðŸ”¹ Removendo duplicatas com base na coordenada mais prÃ³xima
gdf_result = gdf_result.drop_duplicates(subset=['closest_lat', 'closest_lon'], keep='first').reset_index(drop=True)

# Mostrar resultados
print(gdf_result)
```

```python
# ðŸ”½ Salvando o resultado em um arquivo TXT
output_txt_path = '/content/drive/MyDrive/Movasc2_scripts/dist_euclid2.txt'

# Formatando cada linha: lat_in lon_in -> lat_out lon_out
with open(output_txt_path, 'w') as f:
    f.write("LAT_IN\tLON_IN\tCLOSEST_LAT\tCLOSEST_LON\n")
    for i, row in gdf_result.iterrows():
        f.write(f"{row['lat']:.6f}\t{row['lon']:.6f}\t{row['closest_lat']:.6f}\t{row['closest_lon']:.6f}\n")

print(f"Resultado salvo em: {output_txt_path}")
```
